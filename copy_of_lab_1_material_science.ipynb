```python
import numpy as np

# Define lattice vectors for diamond structure
a = 3.43 # lattice constant in Angstroms
lattice_vectors = np.array([[0.5*a, 0.5*a, 0], [0, 0.5*a, 0.5*a], [0.5*a, 0, 0.5*a]])

# Define basis vectors for diamond structure
basis_vectors = np.array([[0, 0, 0], [0.25*a, 0.25*a, 0.25*a]])

# Create a function to generate a diamond unit cell
def diamond_unit_cell():
    # Generate a list of atom positions in the unit cell
    atom_positions = []
    for i in range(3):
        for j in range(3):
            for k in range(3):
                for basis_vector in basis_vectors:
                    atom_position = i*lattice_vectors[0] + j*lattice_vectors[1] + k*lattice_vectors[2] + basis_vector
                    atom_positions.append(atom_position)
    return atom_positions

# Create three layers of the diamond structure
layer_1 = diamond_unit_cell()
layer_2 = [atom_position + lattice_vectors[2] for atom_position in layer_1]
layer_3 = [atom_position + 2*lattice_vectors[2] for atom_position in layer_1]

# Write positions of atoms in each layer to a .xyz file
with open('diamond.xyz', 'w') as f:
    # Write number of atoms in each layer
    f.write(f"{len(layer_1)}\n\n")

    # Write atom positions for layer 1
    for atom_position in layer_1:
        f.write(f"C {atom_position[0]} {atom_position[1]} {atom_position[2]}\n")

    # Write atom positions for layer 2
    for atom_position in layer_2:
        f.write(f"C {atom_position[0]} {atom_position[1]} {atom_position[2]}\n")

    # Write atom positions for layer 3
    for atom_position in layer_3:
        f.write(f"C {atom_position[0]} {atom_position[1]} {atom_position[2]}\n")

# Print the positions of atoms in each layer
print("Layer 1:")
for atom_position in layer_1:
    print(atom_position)

print("\nLayer 2:")
for atom_position in layer_2:
    print(atom_position)

print("\nLayer 3:")
for atom_position in layer_3:
    print(atom_position)
```
